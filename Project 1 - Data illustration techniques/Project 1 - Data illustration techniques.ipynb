```python
# Scatter plot using pyplot library
import matplotlib.pyplot as plt
temperature=[14.2, 16.4, 11.9, 12.5, 18.9, 22.1, 19.4, 23.1, 25.4, 18.1, 22.6, 17.2]
sales=[215.20, 325.00, 185.20, 330.20,418.60,520.25,412.20,614.60,544.80,421.40,445.50,408.10]
plt.scatter(temperature,sales,color='red')
plt.title('awesome')
plt.xlabel('temp')
plt.ylabel('sales')
plt.scatter(temperature,sales,color='red')
plt.show()
```

```python
# Plot diagram
stock_price=[190.64,190.09,192.25,191.79,194.45,196.45,196.45,196.42,200.32,200.32,200.85,199.2,199.2,199.2,200,201,206,210,211,195,196,199,205,210,222,230,203,190,194,206]
t=list(range(1,31))
plt.plot(t,stock_price,marker='.',color='red')
plt.xticks([1,8,15,22,28])
plt.title('Stock prices for TSLA')
plt.xlabel('Time')
plt.ylabel('Value')
```

```python
# Bar plot
grades=['A','B','C','D','E','F']
students_counts=[20,30,10,5,8,2]
plt.title('Student grades distribution')
plt.xlabel('grades')
plt.ylabel('students')
plt.bar(grades,students_counts,color=['green','grey','grey','grey','grey','red'])
```

```python
# Pie chart
labels=['M','A','N']
num=[230,100,40]
plt.title('Voting results: club president',fontdict={'fontsize':30})
plt.pie(num, labels=labels, autopct='%1.1f%%',colors=['yellow','green','blue'])
```

```python
# Bar plot using Seaborn library
import seaborn as sns
data=[90,80,50,42,89,78,34,70,67,73,74,80,60,90,90]
fig=sns.displot(data)
fig.set(title= "Density plot",xlabel = 'Score', ylabel = 'Density')
```
